# Expt-1: Breast Cancer

```python
import matplotlib.pyplot as plt

x_values = [0.1 * i for i in range(8, 13)]  # 10 values from 0.1 to 1.0
y_values1 = [92.9825, 93.8596, 92.9825, 90.3509, 91.2281]
y_values2 = [92.9825, 93.8596, 93.8596, 92.9825, 92.1053]

# Plot the data
plt.plot(x_values, y_values1, linestyle='-', label='RVI', marker='o', alpha=0.7, color='blue')  # Marker 'o' for points
plt.plot(x_values, y_values2, linestyle='--',label='SVI', marker='s', alpha=0.7, color='red')  # Marker 's' for square points

# Add labels, title, and legend
plt.xlabel('Gaussian noise level')
plt.ylabel('Accuracy')
plt.title('Model accuracy vs noise in Breast Cancer data')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
# Display the plot
plt.show()
```

```python
# Here are ELBO for gaussian VAE: [-180.2731, -193.3229, -203.3614, -212.1400, -226.1363]
# ELBO for RVI VAE: [-182.3607, -192.5462, -201.3686, -215.8952, -226.5704]

# RVI @ 0.01: [-170.1141, -180.6327, -190.9523, -203.7080, -216.3246]
# VAE: [-168.7697, -179.2173, -190.4616, -203.6955, -216.4224]
import matplotlib.pyplot as plt


x_values = [0.01 * i for i in range(5, 26, 5)]  # 10 values from 0.1 to 1.0
y_values1 = [-182.3607, -192.5462, -201.3686, -215.8952, -226.5704]
y_values2 =  [-180.2731, -193.3229, -203.3614, -212.1400, -226.1363]

# y_values1 = [-170.1141, -180.6327, -190.9523, -203.7080, -216.3246]
# y_values2 = [-168.7697, -179.2173, -190.4616, -203.6955, -216.4224]

# Plot the data
plt.plot(x_values, y_values1, linestyle='-', label='RVI VAE', marker='o', alpha=0.7, color='blue')  # Marker 'o' for points
plt.plot(x_values, y_values2, linestyle='--',label='A-VI VAE', marker='s', alpha=0.7, color='red')  # Marker 's' for square points

# Add labels, title, and legend
plt.xlabel('Gaussian noise level')
plt.ylabel('Negative ELBO')
plt.title('Negative ELBO vs noise in MNIST')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
# Display the plot
plt.show()
```
